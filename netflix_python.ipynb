```python
#filling director null values with unknown

df.fillna({"director":"unknown"},inplace=True)
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
df.info()
```

```python
#converting date_added colomn datatype to datetime 
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df.info()
```

```python
#fixing country column
df['country']=df['country'].str.strip()
```

```python
# creating new column year_added from date_added to know in which year it has been added into netflix
df['year_added'] = df['date_added'].dt.year
```

```python
df['genre'] = df['listed_in'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
```

```python
df.head()
```

```python
#counting types of mobies and shows 
df['type'].value_counts()
```

```python
# counting how many movies were released in each year 
df['release_year'].value_counts().sort_index()
```

```python
#counting top 10 countries producing movies from all over the world
df['country'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,5))
size=df['type'].value_counts()
plt.pie(size)
plt.grid(True)
plt.title("Pie chart of type")
plt.xlabel("shows")
plt.ylabel("movies")
plt.tight_layout()
plt.show()
```

```python
# removed a column called unnamed 
df.drop(columns=['Unnamed: 12'], inplace=True)
```

```python
#to save this cleaned dataframe into device 
df.to_csv("cleaned_netflix_data.csv",index=False)
```
